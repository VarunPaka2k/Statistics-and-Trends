## 3. Violin Plots

```python
def plot_individual_voilinplots(df):
    """
    Plot individual voilin plots for each selected column in the Data.
    """
    selected_columns = ["Life Ladder", "Log GDP per capita"]
    df = df[selected_columns]
    
    # Looping through each selected column and create a violin plot
    for col in df.columns:
        plt.figure(figsize=(3, 3))
        sns.violinplot(y=df[col])
        plt.title(f'Violin Plot of {col}')
        plt.ylabel(col)
        plt.grid(True)
        plt.show()
```

```python
plot_individual_voilinplots(data)
```

## 4. LINE CHART

```python
# Life Ladder: A subjective measure of well-being, also known as the Cantril ladder. 
# People are asked to rate their current lives on a scale from 0 to 10, with 10 being the best possible life and 0 the worst.

# now extracting the top 5 current(2023) happiest countries and the trend that follows.

data23 = data[data['year'] == 2023]
# Now sorting the data in descending order on 'Life Ladder'
sorted_data = data23.sort_values(by='Life Ladder',ascending = False)
```

```python
# the following countries has highest life ladder in 2023
top_5_countries = sorted_data['Country name'].head().unique()
print(top_5_countries)
```

```python
def plot_trend_of_top5_Life(df):
    """
    Function to plot Life Ladder trends for the top 5 countries / bottom 5 countries over the years.
    """
    plt.figure(figsize=(16, 8))
    
    # Ploting Life Ladder for each country in top 5
    for country in top_5_countries:
        country_data = df[df['Country name'] == country]
        plt.plot(country_data['year'], country_data['Life Ladder'],linestyle='--' ,label=country, marker='o', markersize=4)
    
    # Adding labels and title
    plt.title('Life Ladder Trends for Top 5 Countries in 2023', fontsize=16)
    plt.xlabel('Year', fontsize=15)
    plt.ylabel('Life Ladder', fontsize=15)
    plt.legend(title='Country', loc='upper left')
    plt.grid(linestyle='--')
    
    # Adjusting layout and showing the plot
    plt.tight_layout()
    plt.show()
```

```python
# Calling the function to plot the data.
plot_trend_of_top5_Life(data)
```

## 5. PIE CHART

```python
# Defining happiness level categories and filter the data for each level
low_data = data23[(data23['Life Ladder'] >= data23['Life Ladder'].min()) & (data23['Life Ladder'] < 4)]
medium_data = data23[(data23['Life Ladder'] >= 4) & (data23['Life Ladder'] < 6)]
high_data = data23[(data23['Life Ladder'] >= 6) & (data23['Life Ladder'] <= data23['Life Ladder'].max())]
```

```python
# Calculateing the count for each happiness levels
low_count = low_data.shape[0]
medium_count = medium_data.shape[0]
high_count = high_data.shape[0]
    
# To Create the pie chart using the counts
happiness_distribution = [low_count, medium_count, high_count]
labels = ["Low", "Medium", "High"]
```

```python
def happiness_level_pie_chart(data, score_column):
    """
    Plots a pie chart showing the distribution of happiness levels (High, Medium, Low).
    """
    plt.figure(dpi = 144)
    plt.pie(happiness_distribution , labels = labels, autopct = '%1.1f%%', startangle = 140)
    plt.title('Proportion of Countries by Happiness Level')
    plt.show()
```
